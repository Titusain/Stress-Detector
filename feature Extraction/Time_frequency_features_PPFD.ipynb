```python
def arithmatic_features(window):
  features = []
  features.append(np.mean(window))
  features.append(np.var(window))
  features.append(np.std(window))

  features.append(np.min(window))
  features.append(np.max(window))
  features.append(features[4] - features[3])
  features.append(np.median(window))
  data = window
  var = np.var(data)
  features.append(var)
  data_1 = np.diff(data)
  data_2 = np.diff(data_1)
  hj_mb = np.sqrt(np.var(data_1) / var)
  hj_comp = (np.sqrt(np.var(data_2) / np.var(data_1))) / hj_mb
  features.append(hj_mb)
  features.append(hj_comp)
  return features
```

```python
def Generate_IMF(df):
  w,h = df.shape
  IMF1 = [[0 for x in range(w)] for y in range(h)]
  IMF1=np.array(IMF1)
  IMF2 = [[0 for x in range(w)] for y in range(h)]
  IMF2=np.array(IMF2)
  IMF3 = [[0 for x in range(w)] for y in range(h)]
  IMF3=np.array(IMF3)
  IMF4 = [[0 for x in range(w)] for y in range(h)]
  IMF4=np.array(IMF4)
  IMF5 = [[0 for x in range(w)] for y in range(h)]
  IMF5=np.array(IMF5)
  emd = EMD(DTYPE=np.float16)
  number = -1
  for i in df:
    number+=1
    self = df[i].to_numpy()
    IMF = emd.emd(self, max_imf=5)
    N= IMF.shape[0]+1
  number = -1
  # IMF3.shape,IMF4.shape
  for i in df:
    number+=1
    for k in range(len(IMF[0])):
      IMF1[number][k]=IMF[0][k]
    for k in range(len(IMF[1])):
      IMF2[number][k]=IMF[1][k]
    for k in range(len(IMF[2])):
      IMF3[number][k]=IMF[2][k]
    for k in range(len(IMF[3])):
      IMF4[number][k]=IMF[3][k]
    for k in range(len(IMF[4])):
      IMF5[number][k]=IMF[4][k]
  return IMF1,IMF2,IMF3,IMF4,IMF5
```

```python
def tfdomain(df):
  arr=[]
  IMF1,IMF2,IMF3,IMF4,IMF5=Generate_IMF(df)
  for i in range(32):
      IMF_1=IMF1[i]
      IMF_2=IMF2[i]
      IMF_3=IMF3[i]
      IMF_4=IMF4[i]
      IMF_5=IMF5[i];
      for j in range(25):
          IMF__1=IMF_1[j*128:(j+1)*128]
          arr.append(arithmatic_features(IMF__1))
          IMF__2=IMF_2[j*128:(j+1)*128]
          arr.append(arithmatic_features(IMF__2))
          IMF__3=IMF_3[j*128:(j+1)*128]
          arr.append(arithmatic_features(IMF__3))
          IMF__4=IMF_4[j*128:(j+1)*128]
          arr.append(arithmatic_features(IMF__4))
          IMF__5=IMF_5[j*128:(j+1)*128]
          arr.append(arithmatic_features(IMF__5))
  return arr
```

```python
def reset(arr):
  ans=[]
  for i in range(25):
    temp=[]
    for j in range(32):
      for k in range(5):
        for z in range(10):
          # print(j*125+k+5*i,z)
          temp.append(arr[j*125+k+5*i][z])
    ans.append(temp)
  return ans
```
